```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv("/kaggle/input/d/atharvavengurlekar1/titanic/train.csv")

plt.figure(figsize=(10, 6))
sns.histplot(data=data, x="Age", hue="Survived", multiple="stack", kde=True)
plt.title("Age Distribution by Survival Status")
plt.xlabel("Age")
plt.ylabel("Count")
plt.legend(title="Survived", labels=["Not Survived", "Survived"])
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("/kaggle/input/d/atharvavengurlekar1/titanic/train.csv")

survival_by_gender = data.groupby(['Sex', 'Survived']).size().unstack()

fig, axes = plt.subplots(1, 2, figsize=(15, 7))

for i, gender in enumerate(survival_by_gender.index):
    axes[i].pie(survival_by_gender.loc[gender], labels=['Not Survived', 'Survived'], autopct='%1.1f%%', startangle=140,)
    axes[i].set_title(f'Survival Distribution for {gender.capitalize()}')

plt.show()


```

```python
plt.figure(figsize=(6, 3))
sns.scatterplot(data=data, x="Age", y="Fare", hue="Survived")
plt.title("Scatter Plot of Age vs Fare")
plt.xlabel("Age")
plt.ylabel("Fare")
plt.legend(title="Survived")
plt.show()
```
